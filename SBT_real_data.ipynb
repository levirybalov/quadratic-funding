```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('hackathon-contributions-dataset_v2.csv')
```

```python
user_ids_unique = df.user_id.unique()
len(df.user_id.unique()), len(df.address.unique())
```

```python
# method 1
new_df = df.groupby(['user_id', 'grant_id']).aggregate({'amount_in_usdt': ['sum']})
new_df = new_df.unstack('grant_id')#.fillna(0)
new_df
```

```python
# method 2
new_df2 = df.pivot_table(
    values='amount_in_usdt', 
    index=['user_id'], 
    columns='grant_id', 
    aggfunc=np.sum)#.fillna(0)
new_df2
```

```python
# check that these two methods yield the same matrix
# NOTE: this only works if nan is replaced with zero
# np.allclose(new_df, new_df2)
```

```python
# get the number of projects each user contributed to
num_projects_each_user_contributed_to = new_df2.count(axis=1)
```

```python
# get histogram of number of projects the users contributed to 
# (truncated to users who contributed to less than 50 projects)
num_projects_each_user_contributed_to[num_projects_each_user_contributed_to < 50].hist(bins=100);
```
